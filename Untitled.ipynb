```python
df = pd.read_csv('data.csv')
```

### Data Preprocessing

```python
df.drop(['id', 'Unnamed: 32'], axis = 1, inplace = True)
```

```python
df.diagnosis.unique()
```

This code is using the unique() method to return an array of unique values present in the "diagnosis" column of the DataFrame called "df".

```python
df['diagnosis'] = df['diagnosis'].apply(lambda val: 1 if val == 'M' else 0)
```

In this case, the function passed as an argument to apply() is a lambda function that takes one argument, "val", and returns 1 if "val" is equal to "M" (uppercase M) and 0 otherwise. This lambda function is used to map the values of the "diagnosis" column to binary values, 1 or 0.

### Exploratory Data Analysis (EDA)

```python
plt.figure(figsize = (20, 15))
plotnumber = 1

for column in df:
    if plotnumber <= 30:
        ax = plt.subplot(5, 6, plotnumber)
        sns.distplot(df[column])
        plt.xlabel(column)
        
    plotnumber += 1

plt.tight_layout()
plt.show()
```

This code is using the matplotlib library to create a figure with a specified size (20 inches wide and 15 inches tall). It then loops through each column in a DataFrame and plots a histogram of the values in that column using the seaborn library's distplot function. The x-axis label for each plot is set to the name of the column. The subplot function is being used to arrange the plots in a grid with 5 rows and 6 columns. The tight_layout() function is being used to automatically adjust the spacing between subplots. Finally, plt.show() is used to display the figure on the screen.

```python
# heatmap 

plt.figure(figsize = (20, 12))

corr = df.corr()
mask = np.triu(np.ones_like(corr, dtype = bool))

sns.heatmap(corr, mask = mask, linewidths = 1, annot = True, fmt = ".2f")
plt.show()
```

This code is using the matplotlib library to create a figure with a specified size (20 inches wide and 12 inches tall). It then calculates the pairwise correlation between columns in the DataFrame 'df' and stores it in a variable called 'corr'. The numpy library is used to create a 'mask' that is an upper triangle matrix with the same dimensions as 'corr', where the upper triangle part of the matrix is filled with True values and the lower triangle is filled with False values. The seaborn library's heatmap function is then used to plot the correlation matrix, using the mask to only display the upper triangle of the correlation matrix, and annotating the heatmap with the correlation coefficient values rounded to 2 decimal places. Finally, plt.show() is used to display the figure on the screen.

#### We can see that there are many columns which are very highly correlated which causes multicollinearity so we have to remove highly correlated features

```python
# removing highly correlated features

corr_matrix = df.corr().abs() 

mask = np.triu(np.ones_like(corr_matrix, dtype = bool))
tri_df = corr_matrix.mask(mask)

to_drop = [x for x in tri_df.columns if any(tri_df[x] > 0.92)]

df = df.drop(to_drop, axis = 1)

print(f"The reduced dataframe has {df.shape[1]} columns.")
```

This code is creating a correlation matrix of a dataframe, then creating a mask for the matrix to only show the upper triangle of the matrix. It then drops any columns in the original dataframe that have a correlation greater than 0.92 with another column in the dataframe, and prints the number of columns remaining in the dataframe after the columns have been dropped.

```python
# creating features and label 

X = df.drop('diagnosis', axis = 1)
y = df['diagnosis']
```

```python
# splitting data into training and test set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 0)
```

```python
# scaling data

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print(X_train.shape)
print(X_test.shape)
```

## Logistic Regression

```python
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression

# create an instance of the imputer
imputer = SimpleImputer(strategy='mean')

# fit and transform the imputer on the training data
X_train = imputer.fit_transform(X_train)

# transform the test data using the imputer
X_test = imputer.transform(X_test)

# create an instance of the model
log_reg = LogisticRegression()

# fit the model to the training data
log_reg.fit(X_train, y_train)
```

This code is using the 'SimpleImputer' class from the 'sklearn.impute' library to handle missing values in the training and test data, and the 'LogisticRegression' class from the 'sklearn.linear_model' library to fit a logistic regression model to the training data.

```python
y_pred = log_reg.predict(X_test)
```

```python
# accuracy score

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

log_reg_acc = accuracy_score(y_test, log_reg.predict(X_test))
print(log_reg_acc)
```

```python
#Test model
predections= log_reg.predict(X_test)
print(predections)
print(y_test)
```

```python
# confusion matrix

print(confusion_matrix(y_test, y_pred))
```

```python
# classification report

print(classification_report(y_test, y_pred))
```

## Support Vector Classifier (SVC)

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

svc = SVC()
parameters = {
    'gamma' : [0.0001, 0.001, 0.01, 0.1],
    'C' : [0.01, 0.05, 0.5, 0.1, 1, 10, 15, 20]
}

grid_search = GridSearchCV(svc, parameters)
grid_search.fit(X_train, y_train)
```

'SVC' is a class for fitting a support vector classifier (SVC) model.
'svc' is an instance of the SVC class, without any parameters.
'parameters' is a dictionary that contains a list of possible values for the hyperparameters 'gamma' and 'C' that will be used in the grid search.
'GridSearchCV' is a class that performs an exhaustive search over specified parameter values for an estimator.
'grid_search' is an instance of the GridSearchCV class, initialized with the svc and parameters arguments.

```python
# best parameters

grid_search.best_params_
```

```python
# best accuracy 

grid_search.best_score_
```

```python
svc = SVC(C = 10, gamma = 0.01)
svc.fit(X_train, y_train)
```

```python
# model predictions 

y_pred = svc.predict(X_test)
```

```python
#Printing predictions and actual values
print(y_pred)
print(y_test)
```

```python
# accuracy score
svc_acc = accuracy_score(y_test, svc.predict(X_test))
print(svc_acc)
```

```python
# confusion matrix

print(confusion_matrix(y_test, y_pred))
```

```python
# classification report

print(classification_report(y_test, y_pred))
```

## Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()

parameters = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : range(2, 32, 1),
    'min_samples_leaf' : range(1, 10, 1),
    'min_samples_split' : range(2, 10, 1),
    'splitter' : ['best', 'random']
}

grid_search_dt = GridSearchCV(dtc, parameters, cv = 5, n_jobs = -1, verbose = 1)
grid_search_dt.fit(X_train, y_train)
```

```python
# best parameters

grid_search_dt.best_params_
```

```python
# best score

grid_search_dt.best_score_
```

```python
dtc = DecisionTreeClassifier(criterion = 'entropy', max_depth = 28, min_samples_leaf = 1, min_samples_split = 8, splitter = 'random')
dtc.fit(X_train, y_train)
```

```python
y_pred = dtc.predict(X_test)
```

```python
# accuracy score
dtc_acc = accuracy_score(y_test, dtc.predict(X_test))
print(dtc_acc)
```

```python
# confusion matrix

print(confusion_matrix(y_test, y_pred))
```

```python
# classification report

print(classification_report(y_test, y_pred))
```
